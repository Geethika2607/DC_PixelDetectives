```python
import requests
import urllib.parse  # For URL encoding

# Function to fetch product details from Google Shopping API via RapidAPI
def fetch_google_products(keyword, country="uk", language="en", page=1, limit=5):
    encoded_keyword = urllib.parse.quote(keyword)
    
    url = (
        f"https://real-time-product-search.p.rapidapi.com/search"
        f"?q={encoded_keyword}&country={country}&language={language}"
        f"&page={page}&limit={limit}&sort_by=BEST_MATCH"
        f"&product_condition=ANY&min_rating=ANY"
    )

    headers = {
        "x-rapidapi-key": "8b7b75a47dmsh02d3eacd1f92a9cp1d74bdjsn4f637a6d6180",
        "x-rapidapi-host": "real-time-product-search.p.rapidapi.com"
    }

    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
        data = response.json()
    except requests.exceptions.RequestException as e:
        print("❌ API request error:", e)
        return
    except ValueError:
        print("❌ Error decoding JSON response")
        return

    products = data.get("data", {}).get("products", [])

    if not products:
        print("❌ No products found.")
        return

    print("\n🛍️ Google Shopping Search Results:\n")
    for product in products[:limit]:
        name = product.get("product_title", "Unknown Product")
        price_range = product.get("typical_price_range", ["N/A"])
        price = price_range[0] if price_range else "N/A"
        rating = product.get("product_rating", "N/A")
        num_reviews = product.get("product_num_reviews", "N/A")

        attributes = product.get("product_attributes", {})
        size = attributes.get("Size", "Unknown")
        material = attributes.get("Material", "Unknown")
        color = attributes.get("Color", "Unknown")

        buy_url = product.get("product_page_url", "URL Not Available")
        offers_url = product.get("product_offers_page_url", "URL Not Available")
        images = product.get("product_photos", ["No Image Available"])
        image = images[0] if images else "No Image Available"

        print(f"🔹 Name: {name}")
        print(f"💰 Price: {price}")
        print(f"⭐ Rating: {rating} ({num_reviews} reviews)")
        print(f"🎨 Color: {color}, 🧵 Material: {material}, 📏 Size: {size}")
        print(f"🖼 Image: {image}")
        print(f"🔗 Buy Now: {buy_url}")
        print(f"📦 Offers Page: {offers_url}\n")

# Example usage
fetch_google_products("Asus Laptop")
```
