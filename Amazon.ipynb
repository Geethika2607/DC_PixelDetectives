```python
################################### Amazon API ################################

import requests
import urllib.parse
import json

def fetch_amazon_products(keyword):
    encoded_query = urllib.parse.quote(keyword)
    url = f"https://real-time-amazon-data.p.rapidapi.com/search?query={encoded_query}&country=US&limit=20"

    headers = {
        "x-rapidapi-key": "37cedf8febmshbd85ad68906c31ep1aca7ajsn11f438160bb4",  
        "x-rapidapi-host": "real-time-amazon-data.p.rapidapi.com"
    }

    try:
        response = requests.get(url, headers=headers, timeout=10)
        #print("Status Code:", response.status_code)  # Debug
        response.raise_for_status()
        data = response.json()
        
        #print("📦 Raw JSON Response:\n")
        #print(json.dumps(data, indent=2))  # Pretty print

        products = data.get("data", {}).get("products", [])
        if not products:
            print("❌ No products found.")
            return

        print("\n🛒 Amazon Search Results:\n")
        for product in products[:5]:
            title = product.get("product_title", "N/A")
            price = product.get("product_price", "N/A")
            rating = product.get("product_star_rating", "N/A")
            product_url = product.get("product_url", "N/A")
            image_url = product.get("product_photo", "No Image Available")

            print(f"👜 Title: {title}")
            print(f"💰 Price: {price}")
            print(f"⭐ Rating: {rating}")
            print(f"🔗 URL: {product_url}")
            print(f"🖼️ Image: {image_url}\n")

    except requests.exceptions.RequestException as e:
        print("❌ Request error:", e)

# Example Test
fetch_amazon_products("Gray Hewlett-Packard Laptop")
```
