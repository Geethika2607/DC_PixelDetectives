```python
######################## IKEA ###############################

import requests
import urllib.parse  # For URL encoding

# Function to fetch IKEA products based on keyword
def fetch_ikea_products(keyword, country="us", language="en"):
    encoded_keyword = urllib.parse.quote(keyword)
    url = (
        f"https://ikea-api.p.rapidapi.com/keywordSearch"
        f"?keyword={encoded_keyword}&countryCode={country}&languageCode={language}"
    )

    headers = {
        "x-rapidapi-key": "37cedf8febmshbd85ad68906c31ep1aca7ajsn11f438160bb4",
        "x-rapidapi-host": "ikea-api.p.rapidapi.com"
    }

    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
        products = response.json()
    except requests.exceptions.RequestException as e:
        print("âŒ API request error:", e)
        return
    except ValueError:
        print("âŒ Error decoding JSON response")
        return

    if not isinstance(products, list) or len(products) == 0:
        print("âŒ No products found.")
        return

    print("\nğŸ›‹ï¸ IKEA Search Results:\n")
    for product in products[:5]:  # Show only first 5
        name = product.get("name", "Unknown Product")
        type_name = product.get("typeName", "Unknown Type")
        price_info = product.get("price", {})
        price = price_info.get("currentPrice", "N/A")
        image = product.get("image", "No Image Available")
        url = product.get("url", "URL Not Available")

        print(f"ğŸ”¹ Name: {name}")
        print(f"ğŸ“Œ Type: {type_name}")
        print(f"ğŸ’° Price: ${price}")
        print(f"ğŸ–¼ Image: {image}")
        print(f"ğŸ”— URL: {url}\n")

# Example usage
fetch_ikea_products("Office Chair")
```
