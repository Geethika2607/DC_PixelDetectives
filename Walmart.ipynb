```python
#################### Walmart #############################

import requests
import urllib.parse  # For URL encoding

# Function to fetch Walmart products based on keyword
def fetch_walmart_products(keyword):
    encoded_keyword = urllib.parse.quote(keyword)
    url = (
        f"https://realtime-walmart-data.p.rapidapi.com/search"
        f"?keyword={encoded_keyword}&page=1&sort=price_high"
    )

    headers = {
        "x-rapidapi-key": "37cedf8febmshbd85ad68906c31ep1aca7ajsn11f438160bb4",
        "x-rapidapi-host": "realtime-walmart-data.p.rapidapi.com"
    }

    try:
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
        data = response.json()
    except requests.exceptions.RequestException as e:
        print("‚ùå API request error:", e)
        return
    except ValueError:
        print("‚ùå Error decoding JSON response")
        return

    products = data.get("results", [])

    if not products:
        print("‚ùå No products found in the API response.")
        return

    print("\nüõí Walmart Search Results:\n")
    for product in products[:5]:  # Show only first 5 products
        title = product.get("name", "N/A")
        price = product.get("price", "N/A")
        rating = product.get("rating", "N/A")
        url = product.get("canonicalUrl", "N/A")
        image = product.get("image", "N/A")
        
        if price != "N/A":
            print(f"üîπ Title: {title}")
            print(f"üí∞ Price: ${price}")
            print(f"‚≠ê Rating: {rating}")
            print(f"üîó URL: {url}")
            print(f"üñºÔ∏è Image: {image}\n")

# Example usage
fetch_walmart_products("Gray Hewlett-Packard Laptop")
```
